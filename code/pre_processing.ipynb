```python
def check_empty_img(url):
    # Reading Image
    # You can give path to the 
    # image as first argument
#     print(url)
    url_components = url.split('/')
    
    # Get the last component, which should be the image filename
    image_filename = url_components[-1]
    
    # Check if the image filename contains 'B01' and not '1_B01'
    if 'B01' in image_filename:
        return True
    else:
        return False
```

```python
def stack_bands(path):
    print(path)

    all_files = os.listdir(path)
    band_list = [file for file in all_files if file.endswith('B02.tif') or file.endswith('B03.tif') or file.endswith('B04.tif') or file.endswith('B08.tif')]

    try:
        
        # Read metadata of first file
        with rasterio.open(path +'/'+ band_list[0]) as src0:
            meta = src0.meta

        # Update meta to reflect the number of layers
        meta.update(count = len(band_list))

        # Read each layer and write it to stack
        with rasterio.open(path + '/' + 'stack.tif', 'w', **meta) as dst:
            for id, layer in enumerate(band_list, start=1):
                with rasterio.open(path +'/'+ layer) as src1:
                    dst.write_band(id, src1.read(1))
    except:
        print("Folder with no Data")
        remove_empty_folders(path)
        
        pass
```
